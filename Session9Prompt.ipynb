```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Pulls 1000 random numbers distributed uniformly between [0,1]
randomNums = np.random.uniform(0, 1, 1000)

# Creates a histogram with 100 bins
plt.hist(randomNums, bins=100)

# Label axes
plt.xlabel('X')
plt.ylabel('Y')

# Save as a PDF
plt.savefig('randomNumsHistogram.pdf', format='pdf')

# Show the plot
plt.show()
```
