```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Pulls 1000 random numbers from a normal distribution
random_nums = np.random.normal(size=1000)
```

```python
# Creates a histogram with 100 bins
plt.hist(random_nums, bins=100, density=True)  # `density=True` to normalize the histogram

# Label axes
plt.xlabel('X')
plt.ylabel('Y')

# Save as a PDF file
plt.savefig('normal_distribution_histogram.pdf', format='pdf')

# Display the plot
plt.show()
```
