```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creates an array x ranging from [0,1] in 100 steps
x = np.linspace(0, 1, 100)

# Defines sin
def sinFunct(x):
    return np.sin(x)

# Defines cos
def cosFunct(x):
    return np.cos(x)
```

```python
# Calculates y values using the sin and cos
ySin = sinFunct(x)
yCos = cosFunct(x)

# Creates a figure and a set of subplot
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot sin(x) vs x in the left panel
axs[0].plot(x, ySin)
axs[0].set_title('sin(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')
axs[0].grid(True)

# Plot cos(x) vs x in the right panel
axs[1].plot(x, yCos)
axs[1].set_title('cos(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')
axs[1].grid(True)

# Adjust the layout so there's no overlap
plt.tight_layout()

# Save as a PDF
plt.savefig('sinCosPlots.pdf', format='pdf')

# Show the plots
plt.show()
```
