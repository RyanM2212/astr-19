```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creates an array x ranging from [0,1] in 100 steps
x = np.linspace(0, 1, 100)

# Defines the function that returns exp(x)
def expFunction(x):
    return np.exp(x)
```

```python
# Calculates y using the function
y = expFunction(x)

# Plot the function
plt.plot(x, y)
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.grid(True)

# Saves as a PDF
plt.savefig('expFunctPlot.pdf', format='pdf')

# Show the plot
plt.show()
```
